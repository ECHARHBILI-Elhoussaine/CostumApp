```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.application import MIMEApplication
import time
import os

# ===== CONFIGURATION =====
SMTP_SERVER = "smtp.gmail.com"  # Ou "smtp.office365.com", etc.
SMTP_PORT = 587
EMAIL_FROM = ".....@ump.ac.ma"
EMAIL_PASSWORD = "Voire youtub vedio"  # Pour Gmail : mot de passe d'application : https://youtu.be/MkLX85XU5rU?si=Wq8QWXyv8eWU13Ho
EMAIL_SUBJECT = " Candidature Stage PFE 2026 "

# Corps du message (personnalisable)
EMAIL_BODY = """

Bonjour Madame/Monsieur, j'espère que vous allez bien,
Actuellement en troisième année du cycle d’ingénieur en Génie Informatique à l'Ecole Nationale des Sciences Appliquées d'Oujda    - ENSAO (Université Mohammed Premier, Oujda), je suis à la recherche d’un 𝐬𝐭𝐚𝐠𝐞 de 𝐟𝐢𝐧 𝐝’𝐞𝐭𝐮𝐝𝐞𝐬 - 𝐏𝐅𝐄 d’une durée de 6 mois, à partir de février 2026.
Passionné par le 𝐝𝐞𝐯𝐞𝐥𝐨𝐩𝐩𝐞𝐦𝐞𝐧𝐭 𝐥𝐨𝐠𝐢𝐜𝐢𝐞𝐥 (web, mobile et desktop)..........
"""

# Chemin vers le PDF à attacher (modifiez-le)
PDF_PATH = "/content/******.pdf"  # Ex: "C:/Users/You/CV.pdf" ou "/home/user/cv.pdf , cliquer droit puis chemin"

# Liste des destinataires (1 email = 1 envoi séparé)
DESTINATAIRES = [

                 #EMAILS
]

# ===== FONCTION D'ENVOI =====
def send_emails():
    try:
        # Vérifier si le PDF existe
        if not os.path.exists(PDF_PATH):
            print(f"❌ Erreur : Le fichier '{PDF_PATH}' est introuvable.")
            return

        # Connexion au serveur SMTP
        server = smtplib.SMTP(SMTP_SERVER, SMTP_PORT)
        server.starttls()
        server.login(EMAIL_FROM, EMAIL_PASSWORD)

        for dest in DESTINATAIRES:
            # Création du message
            msg = MIMEMultipart()
            msg['From'] = EMAIL_FROM
            msg['To'] = dest  # Chaque destinataire ne voit que son email
            msg['Subject'] = EMAIL_SUBJECT

            # Ajout du corps du message
            msg.attach(MIMEText(EMAIL_BODY, 'plain'))

            # Ajout du PDF en pièce jointe
            with open(PDF_PATH, "rb") as pdf_file:
                pdf_attachment = MIMEApplication(pdf_file.read(), _subtype="pdf")
                pdf_attachment.add_header(
                    'Content-Disposition',
                    'attachment',
                    filename=os.path.basename(PDF_PATH)  # Nom du fichier dans l'email
                )
                msg.attach(pdf_attachment)

            # Envoi
            server.sendmail(EMAIL_FROM, dest, msg.as_string())
            print(f"✅ Email envoyé à {dest} avec le PDF attaché")

            # Délai anti-spam (2-5 secondes)
            time.sleep(5)

        server.quit()
        print("🎉 Tous les emails ont été envoyés avec succès !")

    except Exception as e:
        print(f"❌ Erreur : {str(e)}")

# Lancement
if __name__ == "__main__":
    send_emails()
```
